```python
from faker import Faker
import pandas as pd
import numpy as np
import random
```

```python
fake = Faker()
```

```python
tranDate = []
custName = []
cardNum = []
zipCode = []
tranAmount = []
prodId = []
prodCate = []
brand = []
starRate = []
prodColor = []
size = []
mrp = []
sellPrice = []
discount = []
custId = []
quant = []
orderId = []
age = []
city = []

brand_names = [ "Adidas", "Puma", "Reebok", "Under Armour", "Levi's", "Gucci", "Prada", "Zara", "H&M"]


product_categories = [
    "Tops", "Dresses", "Bottoms", "Outerwear", "Lingerie"
]

size_options = {
    "Tops": ["S", "M", "L", "XL", "XXL"],
    "Dresses": ["S", "M", "L", "XL", "XXL"],
    "Bottoms": ["S", "M", "L", "XL", "XXL"],
    "Outerwear": ["S", "M", "L", "XL", "XXL", "XS", "XXXL"],
    "Lingerie": ["S", "M", "L", "XL", "XXL"]
}

mrp_ranges = {
      "Zara": (500, 4000),
      "H&M": (400, 3000),
      "Gucci": (800, 4000),
      "Prada": (500, 3000),  # Added
      "Reebok": (800, 4000),   # Added
      "Under Armour": (2500, 3000),  # Added
      "Adidas": (500, 4000),  # Added
      "Puma": (600, 3000),     # Added
      "Levi's": (400, 2000)    # Added
}


def get_discount(mrp):
    if mrp > 5000:
        return 5  # 5% discount
    elif mrp > 3000:
        return 4  # 4% discount
    elif mrp > 1000:
        return 3  # 3% discount
    else:
        return 2

```

```python
def dummy_data(numRec):
    for _ in range(numRec):
        tranDate.append(fake.date_time_between_dates('-1y', 'now'))
    for _ in range(numRec):
        custName.append(fake.name())
    for _ in range(numRec):
        cardNum.append(fake.credit_card_number())
    for _ in range(numRec):
        zipCode.append(fake.zipcode())

    for _ in range(numRec):
        prodId.append(fake.random.randint(101, 500))
    for _ in range(numRec):
        prodCate.append(random.choice(product_categories))

    for _ in range(numRec):
        selected_brand = random.choice(list(mrp_ranges.keys()))  # Ensure brand exists in mrp_ranges
        brand.append(selected_brand)

    for _ in range(numRec):
        starRate.append(fake.random.randint(1, 5))
    for _ in range(numRec):
        prodColor.append(fake.color_name())
    for _ in range(numRec):
        category = random.choice(product_categories)
        size.append(random.choice(size_options[category]))

    for _ in range(numRec):
      selected_brand = random.choice(list(mrp_ranges.keys()))  # Ensure brand exists in mrp_ranges
      brand.append(selected_brand)
      min_mrp, max_mrp = mrp_ranges[selected_brand]  # Fetch correct MRP range
      item_mrp = random.randint(min_mrp, max_mrp)
      mrp.append(item_mrp)  # Store the generated MRP for each record

    for i in range(numRec):
      discount_percent = get_discount(mrp[i])  # Get discount for the individual MRP
      discount.append(discount_percent)

    for i in range(numRec):
      item_sell_price = mrp[i] - (mrp[i] * discount[i] / 100)  # Use the individual MRP and discount
      sellPrice.append(round(item_sell_price, 2))  # Store the calculated sell price


    for _ in range(numRec):
        custId.append(fake.random.randint(1, 5000))

    for _ in range(numRec):
        quant.append(fake.random.randint(1, 10))
    for _ in range(numRec):
        orderId.append(fake.random.randint(1, 500))
    for _ in range(numRec):
        age.append(fake.random.randint(15, 70))
    for _ in range(numRec):
        city.append(fake.city())

    for i in range(numRec):
        final_amount = sellPrice[i] * quant[i]
        tranAmount.append(round(final_amount, 2))
```

```python
tranDate = []
custName = []
cardNum = []
zipCode = []
tranAmount = []
prodId = []
prodCate = []
brand = []
starRate = []
prodColor = []
size = []
mrp = []
sellPrice = []
discount = []
custId = []
quant = []
orderId = []
age = []
city = []

dummy_data(30000)
df = pd.DataFrame(

    zip(custId,custName,age,city,zipCode,orderId,tranDate,cardNum,prodId,prodCate,prodColor,
        brand,size,quant,starRate,mrp,discount,sellPrice,tranAmount),
    columns=['custId','custName','age','city','zipCode','orderId','tranDate','cardNum','prodId','prodCate',
             'prodColor','brand','size','quant','starRate','mrp','discount','sellPrice','tranAmount']

)
df.shape
```
